### FunciÃ³n para descomprimir los datos del socket

```python
def lzw_decompress(compressed_str):
    compressed = [ord(c) for c in compressed_str]
    dict_size = 256
    dictionary = {i: chr(i) for i in range(dict_size)}

    result = []
    w = chr(compressed.pop(0))
    result.append(w)

    for k in compressed:
        if k in dictionary:
            entry = dictionary[k]
        elif k == dict_size:
            entry = w + w[0]
        else:
            raise ValueError(f"Bad compressed k: {k}")
        result.append(entry)

        # Add w+entry[0] to the dictionary
        dictionary[dict_size] = w + entry[0]
        dict_size += 1

        w = entry

    return "".join(result)
```

### FunciÃ³n para insertar los datos

```python
def guardar_rayo_sqlite(data_dict):
    import sqlite3
    conn = sqlite3.connect("rayos.db")
    cursor = conn.cursor()

    # Insertar datos bÃ¡sicos (puedes modificar segÃºn lo que te interese)
    cursor.execute("""
    INSERT INTO rayos (time, lat, lon, alt, pol, mds, mcg, status, region, raw_json)
    VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?)
    """, (
        data_dict.get("time"),
        data_dict.get("lat"),
        data_dict.get("lon"),
        data_dict.get("alt"),
        data_dict.get("pol"),
        data_dict.get("mds"),
        data_dict.get("mcg"),
        data_dict.get("status"),
        data_dict.get("region"),
        json.dumps(data_dict)
    ))

    conn.commit()
    conn.close()
```

# Lectura del socket y guardado a SQLite


```python
import json

async def escuchar():
    url = "wss://ws2.blitzortung.org/"
    headers = {
        "Origin": "https://www.blitzortung.org",
        "User-Agent": "Mozilla/5.0",
    }

    session = aiohttp.ClientSession()
    async with session.ws_connect(url, headers=headers) as ws:
        await ws.send_str('{"a":111}')

        async for msg in ws:
            if msg.type == aiohttp.WSMsgType.TEXT:
                try:
                    raw = msg.data
                    texto = lzw_decompress(raw)
                    data = json.loads(texto)
                    guardar_rayo_sqlite(data)
                    print("âœ… Guardado")
                except Exception as e:
                    print("âŒ Error:", e)
            elif msg.type in [aiohttp.WSMsgType.CLOSED, aiohttp.WSMsgType.ERROR]:
                break

    await session.close()

# Si estÃ¡s en Jupyter:
await escuchar()

# Si estÃ¡s en un .py:
# asyncio.run(escuchar())
```
